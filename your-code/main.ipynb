### Setup

```python
import os
import warnings
from dotenv import load_dotenv
from IPython.display import Markdown, display

from langchain.agents import create_agent
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch

warnings.filterwarnings("ignore")
load_dotenv()

required_keys = ["OPENAI_API_KEY", "TAVILY_API_KEY"]
missing_keys = [key for key in required_keys if not os.getenv(key)]

if missing_keys:
    print("Missing API key(s):", ", ".join(missing_keys))
    print("Add them to a .env file or set them as environment variables before running the agent.")

# Tavily is the web-search tool used by the agent.
tavily_search = TavilySearch(
    max_results=5,
    topic="general",
    search_depth="advanced",
)

# gpt-4o-mini keeps the lab cheaper and fast. You can switch to gpt-4o if your instructor asks for it.
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
)

system_prompt = """
You are a careful research assistant.
Use Tavily web search when the question asks for recent or current information.
Return clean Markdown.
Prefer reliable sources such as official company pages, major publications, research institutions, regulators, and reputable technology/healthcare media.
When possible, include source links next to the claims they support.
Be concise, structured, and factual.
"""

agent = create_agent(
    model=llm,
    tools=[tavily_search],
    system_prompt=system_prompt,
)


def run_agent_markdown(prompt: str) -> str:
    """Run the LangChain agent and return only the final assistant message as Markdown text."""
    result = agent.invoke({"messages": [{"role": "user", "content": prompt}]})
    final_message = result["messages"][-1]
    return final_message.content
```

### Exercise 1: AI in Healthcare

Goal: Investigate and summarize the latest advancements in generative AI applied to healthcare in 2025.

- Design a prompt that asks the agent to retrieve the most recent updates.
- Ensure the agent outputs a structured response in Markdown.

```python
# Exercise 1 prompt: AI in Healthcare
prompt_1 = """
Use real-time web search to investigate the latest 2025 advancements in generative AI applied to healthcare.

Return the answer in clean Markdown with these sections:

# Generative AI in Healthcare: 2025 Update

## Executive Summary
- 4 to 6 bullet points with the most important trends.

## Key Advancements
Create a Markdown table with these columns:
- Area
- What changed in 2025
- Why it matters
- Example organizations/products
- Source links

Cover at least these areas if relevant:
- clinical documentation and ambient scribes
- medical imaging
- drug discovery and clinical trials
- patient communication/chatbots
- hospital operations
- regulation, safety, privacy, and risk

## Main Risks and Limitations
Give 4 to 6 bullet points.

## Practical Conclusion
Explain in one short paragraph how generative AI is most likely to create value in healthcare right now.

Use sources from 2025 where possible. Include links.
"""

response_1 = run_agent_markdown(prompt_1)
```

```python
display(Markdown(response_1))
```

### Exercise 2: AI Startups Landscape

Goal: Track 2025’s top emerging AI startups and their innovations.

- Create a prompt that instructs the agent to deliver a clean Markdown summary.
- Tip: Ask for company names, product highlights, and sources.

```python
# Exercise 2 prompt: AI Startups Landscape
prompt_2 = """
Use real-time web search to identify top emerging AI startups in 2025 and summarize their innovations.

Return the answer in clean Markdown with these sections:

# Emerging AI Startups Landscape: 2025

## Short Overview
Explain the main startup trends in 2025 in 4 to 6 bullets.

## Startup Watchlist
Create a Markdown table with 8 to 10 startups and these columns:
- Startup
- Main product or innovation
- Sector
- Why it is notable in 2025
- Funding/adoption signal if available
- Source links

Prefer startups connected to areas such as:
- AI agents
- generative AI infrastructure
- healthcare AI
- robotics or embodied AI
- coding assistants
- enterprise automation
- AI search/research tools

## Patterns Across the Market
Summarize what these startups have in common.

## Final Takeaway
Give a concise conclusion for a non-technical business audience.

Use recent and reliable sources. Include links.
"""

response_2 = run_agent_markdown(prompt_2)
```

```python
display(Markdown(response_2))
```

### Exercice 3: Compare Two Tech Products

```python
# Exercise 3 prompt: Compare Two Tech Products
prompt_3 = """
Use real-time web search to compare OpenAI's ChatGPT Team plan and Anthropic's Claude Pro plan.

Return the answer in clean Markdown with these sections:

# ChatGPT Team vs Claude Pro

## Important Note
Briefly explain whether these two plans target the exact same type of user or not.

## Comparison Table
Create a Markdown table with these rows:
- Target user
- Monthly price
- Annual price if available
- Main models included
- Context window / file handling if available
- Collaboration features
- Admin/security features
- Best use cases
- Main limitations
- Source links

## Review Summary
Summarize what users and reviewers usually praise or criticize about each product.

## Recommendation
Give a practical recommendation for:
- individual student or learner
- small team
- business team using AI daily

Use official pricing pages and reputable reviews where possible. Include links.
"""

response_3 = run_agent_markdown(prompt_3)
```

```python
display(Markdown(response_3))
```
